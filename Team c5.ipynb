First we know that for each time period, we have the same number of null values in lower bond, upper bond, and mean travel time. We made a resonable assumation that those rows are the same. 

```python
barts_hotspots.loc[barts_hotspots['AM Mean Travel Time (Seconds)'].isnull(), "Day of the Week"].value_counts()
```

We dicide to use the average value groupby the origin, destination, and the day of the week for each column to replace each column's null values

```python
grp_for_AMnull = barts_hotspots.groupby(["Origin Movement ID", "Destination Movement ID", "Day of the Week"]).mean().to_dict()
grp_for_AMnull
```

```python
lst = grp_for_AMnull.keys()
lst = list(lst)
lst
```

```python
def cleaning(row):
    origin = row['Origin Movement ID']
    destination = row["Destination Movement ID"]
    day = row["Day of the Week"]
    answer = [origin, destination, day]
    for col in lst:
        if grp_for_AMnull[col][(origin, destination, day)] != np.NaN:
            print(grp_for_AMnull[col][(origin, destination, day)])
            if row[col] == np.NaN:
                answer.append(grp_for_AMnull[col][(origin, destination, day)])
            else:
                answer.append(row[col])
    answer = pd.Series(answer, index = ['origin', 'dest', 'day'] + lst)
    return answer
```

```python
cleaned = barts_hotspots.apply(cleaning, axis=1)
```
